```python
import requests
from lxml import etree
with open('ipip-asn-CN-ipv6.txt', 'w') as f:
    url = 'https://whois.ipip.net/iso/CN'
    # url = 'https://whois.ipip.net/iso/US'
    header = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
        AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'
    }
    html = requests.get(url, headers=header)
    html.encoding = 'utf-8'
    cxtTree = etree.fromstring(html.text, etree.HTMLParser())
    items = cxtTree.xpath('//div/table/tr')
    # print(len(items))
    # print(items[0].xpath('td')[0].xpath('a/text()'))
    for _,item in enumerate(items):
        newasn = item.xpath('td')[0].xpath('a/text()')
        newname = item.xpath('td')[1].xpath('text()')
        # newipv4 = item.xpath('td')[2].xpath('text()')
        newipv6 = item.xpath('td')[3].xpath('text()')
        if newipv6[0].strip() != '0' :
            f.write(newasn[0].strip() + ', ')
            # f.write(newipv4[0].strip() + ', ')
            f.write(newipv6[0].strip() + ', ')
            f.write(newname[0].strip() + '\n')
        print(newasn[0])

        
```
